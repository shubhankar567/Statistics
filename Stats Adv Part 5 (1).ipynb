# Answer 2:


```python
from scipy.stats import chisquare

# observed values
observed_values = [30, 40, 20, 5, 5, 50] # blue, orange, green, yellow, red, brown

# expected values
total = sum(observed_values)
expected_values = [0.2*total, 0.2*total, 0.2*total, 0.1*total, 0.1*total, 0.2*total]

# perform chi-square test
chi2_statistic, p_value = chisquare(observed_values, f_exp=expected_values)

# print results
print(f"Chi2 Statistic: {chi2_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# interpret results
alpha = 0.05
if p_value < alpha:
    print("We reject the null hypothesis.")
else:
    print("We do not have enough evidence to reject the null hypothesis.")
```

The chi-square test is used to determine if there is a significant association between two categorical variables. The null hypothesis is that there is no association between the variables (i.e., the variables are independent).

If the p-value is less than a pre-determined significance level (usually 0.05), we reject the null hypothesis and conclude that there is a significant association between the variables.

In this case, based on the p-value obtained from running this code, you can determine whether or not there is a significant association between the two groups and the outcomes.

# Answer 4:

```python
from scipy.stats import norm

# Define the sample size and number of successes
n = 500
successes = 60

# Calculate the proportion
p = successes / n

# Calculate the standard error
se = (p * (1 - p) / n) ** 0.5

# Calculate the z-score for a 95% confidence interval
z = norm.ppf(0.975)

# Calculate the margin of error
me = z * se

# Calculate the 95% confidence interval
ci_low = p - me
ci_upp = p + me

# Print the results
print(f"The 95% confidence interval for the true proportion is ({ci_low:.3f}, {ci_upp:.3f})")
```

# Answer 6:

```python
import matplotlib.pyplot as plt
from scipy.stats import chi2
import numpy as np

# Define the degrees of freedom
df = 10

# Create an array of x values
x = np.linspace(0, 30, 1000)

# Calculate the probability density function for the chi-square distribution
pdf = chi2.pdf(x, df)

# Plot the chi-square distribution
plt.plot(x, pdf, label=f'Chi-Square Distribution (df={df})')

# Shade the area corresponding to a chi-square statistic of 15
x_fill = np.linspace(15, 30, 1000)
y_fill = chi2.pdf(x_fill, df)
plt.fill_between(x_fill, y_fill, alpha=0.4)

# Label the axes
plt.xlabel('Chi-Square Statistic')
plt.ylabel('Probability Density')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

```python
from scipy.stats import chisquare

# Define the observed frequencies
observed = [55, 45]

# Define the expected frequencies for a fair coin
expected = [50, 50]

# Conduct the chi-square goodness of fit test
chi2, p = chisquare(observed, f_exp=expected)

# Print the results
print(f"Chi-square statistic: {chi2:.2f}")
print(f"p-value: {p:.4f}")

# Check if the p-value is less than the significance level
if p < 0.05:
    print("We reject the null hypothesis and conclude that the coin is biased towards tails.")
else:
    print("We do not have sufficient evidence to reject the null hypothesis and conclude that the coin is fair.")
```

# Answer 11:

```python
from scipy import stats

# Define the sample size, sample mean, and sample standard deviation
n = 30
x_bar = 72
s = 10

# Define the null hypothesis value and the significance level
mu_0 = 70
alpha = 0.05

# Calculate the t-statistic and p-value
t_statistic = (x_bar - mu_0) / (s / (n ** 0.5))
p_value = stats.t.sf(abs(t_statistic), n - 1) * 2

# Check if p-value is less than the significance level
if p_value < alpha:
    print("We reject the null hypothesis. The population mean is significantly different from 70.")
else:
    print("We do not reject the null hypothesis. The population mean is not significantly different from 70.")
```
