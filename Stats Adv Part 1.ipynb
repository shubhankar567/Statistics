# Answer 1:

In probability theory, a **probability density function (PDF)** is a function that describes the relative likelihood for a continuous random variable to take on a given value. The PDF is used to specify the probability of the random variable falling within a particular range of values, as opposed to taking on any one value.

# Answer 2:

There are two main types of probability distributions: **Discrete probability distributions** and **Continuous probability distributions**. Some common probability distributions include the **binomial distribution**, **Poisson distribution**, and **uniform distribution**. Certain types of probability distributions are used in hypothesis testing, including the **standard normal distribution**, the **F distribution**, and **Student’s t distribution**.

# Answer 3:


```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# Define the mean and standard deviation of the normal distribution
mean = 0
std_dev = 1

# Define the point at which to calculate the probability density
x = 1.5

# Calculate the probability density at x
pdf = norm.pdf(x, mean, std_dev)

# Print the result
print(f"The probability density at x = {x} is {pdf}")

# Plot the normal distribution
x_values = np.linspace(-5, 5, 1000)
y_values = norm.pdf(x_values, mean, std_dev)
plt.plot(x_values, y_values)
plt.title(f"Normal Distribution with Mean = {mean} and Standard Deviation = {std_dev}")
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.show()
```

# Answer 4:

A binomial distribution is a discrete probability distribution that describes the number of successes in a fixed number of independent trials, each with only two possible outcomes: success or failure. The probability of success is the same for each trial.

The binomial distribution has the following properties:
- The mean of the distribution is `μ = np`.
- The variance of the distribution is `σ² = np(1-p)`.
- The standard deviation of the distribution is `σ = √np(1-p)` .

Two examples of events where binomial distribution can be applied are:
1. Tossing a coin: If we toss a coin `n` times, the number of heads can be modeled as a binomial distribution with `n` trials and probability of success `p` equal to 0.5 (assuming the coin is fair).
2. Quality control: In a factory, if we randomly select `n` items from a production line, the number of defective items can be modeled as a binomial distribution with `n` trials and probability of success `p` equal to the proportion of defective items in the population.

# Answer 5:

```python
import numpy as np
import matplotlib.pyplot as plt

# Set the random seed for reproducibility
np.random.seed(0)

# Define the parameters of the binomial distribution
n = 10
p = 0.4

# Generate a random sample of size 1000 from the binomial distribution
sample = np.random.binomial(n, p, size=1000)

# Plot a histogram of the sample
plt.hist(sample, density=True)
plt.title(f"Sample from Binomial Distribution with n = {n} and p = {p}")
plt.xlabel("x")
plt.ylabel("Probability")
plt.show()
```

```python
import numpy as np
from scipy.stats import poisson
import matplotlib.pyplot as plt

mu = 10 # mean of the distribution
x = np.arange(poisson.ppf(0.01, mu), poisson.ppf(0.99, mu))
y = poisson.cdf(x, mu)

plt.plot(x, y)
plt.title('Cumulative Distribution Function of a Poisson Distribution')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.show()
```

# Answer 7:
The Binomial and Poisson distributions are two discrete probability distributions that can be used to model the number of occurrences of some event. Both distributions assume that events are independent.

The key difference between the two distributions is that in a Binomial distribution, there is a fixed number of trials (e.g. flip a coin 3 times), while in a Poisson distribution, there could be any number of events that occur during a certain time interval (e.g. how many customers will arrive at a store in a given hour?).

In summary, the Binomial distribution is used to model the number of successes in a fixed number of independent trials, while the Poisson distribution is used to model the number of events that occur in a fixed interval of time or space.

In a Poisson distribution, the mean and variance are both equal to `λ`, where `λ` is the mean number of successes that occur during a specific interval.

In summary, in a Binomial distribution, the mean is greater than the variance, while in a Poisson distribution, the mean and variance are equal.

# Answer 10:
In a normal distribution, the data is distributed symmetrically around the mean. This means that the most frequent data values appear near the mean, while the least frequent data values appear in the tails of the distribution, far from the mean.

In other words, in a normal distribution, the least frequent data values are those that are several standard deviations away from the mean, either to the left (below the mean) or to the right (above the mean) of the distribution.
